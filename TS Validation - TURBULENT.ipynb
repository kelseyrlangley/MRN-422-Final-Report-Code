```python
def laststep(Rey):
        cnt = 0
        sta  = 0
        test = np.zeros(4)
        Station = np.zeros(13)

        for h in range(0,60-8):
            T = Rey[h]
            test[cnt] = T
            cnt += 1
            if cnt == 4:
                Station[sta] = np.mean(test)
                cnt = 0
                sta += 1
        return np.array(Station)

print(laststep(Re4))
```

# Percentage Dev from Mean

```python
from numpy import mean, absolute 
  
#average deviation and percentage deviation from mean
def avedev(data):
        A = np.mean(data)

        sum = 0  # Initialize sum to 0 

        for i in range(len(data)): 
            av = absolute(data[i] - A)   # Absolute value of the differences  
                                         # of each data point and A 
            # Summing all those absolute values 
            sum = sum + av                

        ad = sum / len(data)
        
        perdev = ad/A *100
        return perdev

print(avedev(RE6000))
print(avedev(RE7000))
print(avedev(RE8000))
print(avedev(RE9000))
print(avedev(RE10000))
```

# TF estimate (Average TS)

```python
D = 0.0127
L = 5.5

m = np.array(pd.read_table("{}.txt".format(34)))[:200,79]
mdot = ((np.mean(m)*1000)*25 - 100)/3600
print("mdot =", mdot)

i = np.array(pd.read_table("{}.txt".format(34)))[:200,77]
Ti = (np.mean(i)-0.2942543581515781)/0.9908373802343359 
print("Ti =",Ti)

e = np.array(pd.read_table("{}.txt".format(34)))[:200,78]
Te = (np.mean(e)-0.5179385816684551)/0.9878359665716272
print("Te =", Te) 

As = np.pi*D*L
print(As)

Tb_i = (Ti+Te)/2
print(Tb_i)

def func(file):
    cat = []
    for i in range(0,52):
        templist = np.array(pd.read_table("{}.txt".format(file)))[:200,i]
        ave = np.mean(templist)
        if 0 < ave < 100:
            cat.append(ave)
        else:
            continue
    return cat
    #print(templist)
        #print(ave)

Ts = np.mean(func(34))
print("Ts =", Ts)

Cp_30 = 4178

a = np.log((Ts-Te)/(Ts-Ti))
print(a)

TLM = (Ti-Te)/a
print(TLM)

have = (mdot*Cp_30*(Te-Ti))/(As*TLM) #As is f(x) ie. will change h
print("h_ave =", have)

axial = np.array([0.060,0.100,0.180,0.280,0.480,0.860,1.200,1.900,2.600,3.300,4.000,4.700,5.400])

As_x = []
for x in axial:
    asx = np.pi*D*x
    As_x.append(asx)
#print("As_x =", As_x)

Tfvals = []
for i in As_x:
    Tf_new = Ts - (Ts-Ti)*np.exp(-(have*i)/(mdot*Cp_30))
    Tfvals.append(Tf_new)

print("Tf =", Tfvals)
```

# Nu Correlations

Fluid Properties

```python
def mu(T):
    return (557.82468 + (19.408782*T) + (0.1360459)*(T**2) + (-3.1160832E-4)*(T**3))**(-1)
def k(T):
    return 0.5650285 + (0.0026363895*T) + (-0.00012516934)*(T**1.5) + (-1.5154918E-6)*(T**2) + (-0.0009412945)*(T**0.5)
def Pr(T):
    return (0.074763403 + (0.0029020983*T) + (2.8606181E-5)*(T**2) + (-8.1395537E-8)*(T**3))**-1
def B(T):
    return -6.8785895E-5 + (2.1687942E-5)*T + (-2.1236686E-6)*(T**1.5) + (7.7200882E-8)*(T**2)
def den(T):
    return 999.79684 + 0.068317355*T + -0.010740248*T**2 + 0.00082140905*T**2.5 + -2.3030988E-5*T**3
    
muvals = []
for t in Tfvals:
    MU = mu(t)
    muvals.append(MU)
print(muvals)

kvals = []
for t in Tfvals:
    K = k(t)
    kvals.append(K)
#print(kvals)

prvals = []
for t in Tfvals:
    PR = Pr(t)
    prvals.append(PR)
print(prvals)

bvals = []
for t in Tfvals:
    BT = B(t)
    bvals.append(BT)
#print(bvals)

dens = []
for t in Tfvals:
    DT = den(t)
    dens.append(DT)
#print(dens)

vvals = []
for i in range(0,13):
    v = muvals[i]/dens[i]
    vvals.append(v)
#print(vvals)
```
